```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(pop['총인구수'])
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
pop1 = pop.sort_values("총인구수")
sns.barplot(x=pop1.index, y='총인구수', data=pop1)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
